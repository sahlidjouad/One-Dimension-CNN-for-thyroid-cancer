```python
dataset_path = os.getenv("DATASET_DIR_HDf5")
data = h5py.File(dataset_path, "r")
dataset = pd.read_csv(os.getenv('DATASET_DIRR'), index_col="annot_id")
```

```python
df = None
radiomic_extractor = RadiomicsExtractor()

for row in range(17412):
    image = data["image"][row]
    fram_num = data["frame_num"][row]
    annot_id = data["annot_id"][row]
    mask = data["mask"][row]
    extracted_features = radiomic_extractor.extract(image, mask)

    data_specific_row = {"annot_id": annot_id.decode()}

    for k, v in six.iteritems(extracted_features):
        if "diagnostics" not in k:
            data_specific_row[k] = v

    final_data_frame = pd.DataFrame(data_specific_row, index=[0]).set_index("annot_id")
    if df is None:
        df = final_data_frame
    else:
        df = pd.concat([df, final_data_frame])
```

```python
df =df.join(dataset)
```

```python
df.to_csv("./../dataset/FullDatasetAndWavlet.csv")
```
